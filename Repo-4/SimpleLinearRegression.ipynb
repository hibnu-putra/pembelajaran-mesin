# EDA

```python
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
df['Salary'].plot(kind='kde')

f.add_subplot(1,2,2)
plt.boxplot(df['Salary'])
plt.show()
```

```python
plt.scatter(df['Salary'], df['YearsExperience'])
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.title('Scatter Plot Salary vs Price')
plt.show()
```

```python
df.corr()
```

# Modeling

```python
x = df['Salary'].values.reshape(-1,1)
y = df['YearsExperience'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.score(x_test, y_test)
```

```python
y_prediksi = lin_reg.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediksi, c='r')
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.title('Plot Salary vs YearsExperience')
```

# Prediction

```python
#Prediksi YearsExperience mobil dengan Salary 76000
lin_reg.predict([[76000]])
```
